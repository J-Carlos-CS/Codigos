```python
# Use spark.sql to round up trip_distance to the nearest 0.5
rounded_trip_distance_df = spark.sql("""
SELECT
    VendorID,
    tpep_pickup_datetime,
    tpep_dropoff_datetime,
    passenger_count,
    trip_distance,
    ceil(trip_distance * 2) / 2 as rounded_trip_distance,
    pickup_longitude,
    pickup_latitude,
    RateCodeID,
    store_and_fwd_flag,
    dropoff_longitude,
    dropoff_latitude,
    payment_type,
    fare_amount,
    extra,
    mta_tax,
    tip_amount,
    tolls_amount,
    improvement_surcharge,
    total_amount
FROM Taxi
""")

# Show the DataFrame with the new rounded_trip_distance column
rounded_trip_distance_df.show()
```

```python
# Group by rounded_trip_distance and count the occurrences
trip_count_by_rounded_distance = rounded_trip_distance_df.groupBy("rounded_trip_distance").count()

# Order the results by rounded_trip_distance
trip_count_by_rounded_distance = trip_count_by_rounded_distance.orderBy("rounded_trip_distance")

# Show the results
trip_count_by_rounded_distance.show()
```

```python
# Filter out rounded_trip_distance values greater than 50.0
filtered_trip_count_by_rounded_distance = trip_count_by_rounded_distance.filter(col("rounded_trip_distance") <= 50.0)

# Convert the filtered Spark DataFrame to a Pandas DataFrame
trip_count_by_rounded_distance_df = filtered_trip_count_by_rounded_distance.toPandas()

# Display the Pandas DataFrame
trip_count_by_rounded_distance_df
```

```python
# Create a bar plot using Plotly Express
fig = px.bar(trip_count_by_rounded_distance_df, x='rounded_trip_distance', y='count',
             title='Distribucion de Distancia por Viaje en millas',
             labels={'rounded_trip_distance': 'Distancia Recorrida', 'count': 'Numero de Viajes'},
             color='rounded_trip_distance') # Color bars by passenger count

fig.show()
```
