```python
# iniciamos las dependencias
import pandas as pd
import json
```

```python
# Read the JSON file into a pandas DataFrame, allowing pandas to infer dtypes
try:

    # Leer el JSON como diccionario
    with open('/content/resultados03.json', 'r') as f:
        raw_data = json.load(f)

    # Convertir a DataFrame, usando las claves como columna
    df = pd.DataFrame.from_dict(raw_data, orient='index').reset_index()
    df['index'] = df['index'].astype(str)  # Asegura que no se interprete como número o fecha

except FileNotFoundError:
    print("Error: data.json not found. Please make sure the file exists in the specified path.")
    df = None
```

```python
df
```

```python
# Transpose the DataFrame
if df is not None:
    df = df.T

    # Expand the 'r' and 'p' columns into new columns
    # Safely apply json.loads and then pd.Series, handling potential errors
    def parse_json_safely(json_string):
        if isinstance(json_string, str):
            try:
                return json.loads(json_string)
            except json.JSONDecodeError:
                return {}  # Return an empty dictionary on JSONDecodeError
        elif isinstance(json_string, dict):
             return json_string # Return dictionary as is
        else:
             return {} # Return an empty dictionary for other types
    r_df = df['r'].apply(parse_json_safely).apply(pd.Series)
    p_df = df['p'].apply(parse_json_safely).apply(pd.Series)
    # Concatenate the new columns with the original DataFrame
    df = pd.concat([df.drop(['r', 'p'], axis=1), r_df, p_df], axis=1)

# Display the first few rows of the DataFrame
if df is not None:
    display(df.head())
```

```python
# dffix = df.rename_axis('c')
# dffix.reset_index(inplace=True)
df.rename(columns={'index': 'c'}, inplace=True)
```
