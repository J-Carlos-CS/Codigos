##Leemos todos los datasets con la libreria Pandas

```python
import pandas as pd
import os
import glob
# Listamos todos los ficheron CSVs
files = glob.glob('fire_*.csv')

# Leemos cada fichero y guardamos en un arreglo
dfs = []
for f in files:
    try:
        df = pd.read_csv(f)
        dfs.append(df)
    except Exception as e:
        print(f"Error reading {f}: {e}")

# Unimos todos los datasets en un solo DataFrame.
if dfs:
    all_fire_data = pd.concat(dfs, ignore_index=True)
    print("Se combinaron todos los datasets en uno solo.")
    display(all_fire_data.head())
    display(all_fire_data.info())
else:
    print("Datasets CSV no encontrados.")
```

```python
# revisamos la cantidad de registros o focos de calor
print(f'La cantidad de focos de calor del año  2001 al 2025 es: {len(all_fire_data)}')
```

## Separamos el campo "acq_date" en year, month, day

```python
# Convert 'acq_date' to datetime objects
all_fire_data['acq_date'] = pd.to_datetime(all_fire_data['acq_date'])

# Extract year, month, and day into new columns
all_fire_data['year'] = all_fire_data['acq_date'].dt.year
all_fire_data['month'] = all_fire_data['acq_date'].dt.month
all_fire_data['day'] = all_fire_data['acq_date'].dt.day

display(all_fire_data.head())
display(all_fire_data.info())
```

## Observamos gráficamente el comportamiento de los incendios y quemas a partir de focos de calor

```python
import plotly.express as px

# Agrupación por año y número de focos de calor
fire_incidents_per_year = all_fire_data['year'].value_counts().sort_index()

# Graficamos los resultados
fig = px.line(x=fire_incidents_per_year.index, y=fire_incidents_per_year.values,
              labels={'x':'Year', 'y':'Number of Incidents'},
              title='Número de incidentes de incedios/quemas por año')
fig.show()
```

## Aplicamos una relación espacial de focos de calor del año 2024 con límites de municipios

```python
len(filtered_fire_data)
```

```python
import geopandas as gpd
from shapely.geometry import Point

# Cargamos los municipios en formato shapefile
municipalities_gdf = gpd.read_file("limite_municipal_2004f.shp")

# Tomamos en cuenta todos los filtros previos de taños y satélites

# Creamos un GeoDataFrame
# Asumimos CRS/Datum del shapefile es WGS84 (EPSG:4326)
filtered_fire_data24 = filtered_fire_data[filtered_fire_data['year']==2024]
fire_gdf = gpd.GeoDataFrame(
    filtered_fire_data24,
    geometry=gpd.points_from_xy(filtered_fire_data24.longitude, filtered_fire_data24.latitude),
    crs="EPSG:4326"
)

if fire_gdf.crs != municipalities_gdf.crs:
    fire_gdf = fire_gdf.to_crs(municipalities_gdf.crs)
    print(f"Reproyectamos los FC con el CRC de municipios: {municipalities_gdf.crs}")

# Aplicamos la relación espacial
fire_incidents_with_municipalities = gpd.sjoin(fire_gdf, municipalities_gdf, how="inner", predicate="intersects")

print("relación espacial (geoproceso):")
display(fire_incidents_with_municipalities.head())
display(fire_incidents_with_municipalities.info())
```

## Guardamos el resultado en un CSV

```python
# Save the GeoDataFrame to a CSV file
# You might want to choose a specific path and filename
output_csv_path = 'FC-municipios-Bo-2024.csv'
fire_incidents_with_municipalities.to_csv(output_csv_path, index=False)

print(f"Data successfully saved to {output_csv_path}")
```

## Generamos la gráfica de focos de calor mensual del municipio con mayor cantidad de focos de calor

```python
import plotly.express as px

# Filter the data for the municipality of 'Concepción'
concepcion_fire_data = fire_incidents_with_municipalities[
    fire_incidents_with_municipalities['NOM_DEP'] == 'Beni'
].copy()

# Group by year and month and count the incidents
concepcion_monthly_incidents = concepcion_fire_data.groupby(['year', 'month']).size().reset_index(name='count')

# Create a mapping for month numbers to abbreviated names in Spanish
month_names = {
    1: 'ene', 2: 'feb', 3: 'mar', 4: 'abr', 5: 'may', 6: 'jun',
    7: 'jul', 8: 'ago', 9: 'sep', 10: 'oct', 11: 'nov', 12: 'dic'
}
# Map the month numbers to names
concepcion_monthly_incidents['month_name'] = concepcion_monthly_incidents['month'].map(month_names)

# Plot the results
fig = px.line(concepcion_monthly_incidents, x='month_name', y='count', color='year',
              labels={'month_name':'Mes', 'count':'Número de focos de calor', 'year':'Año'},
              title='Número de incidentes de incendios mensual en Concepción (2018-2025)') # Assuming the data is for 2018-2025 based on previous filtering
fig.show()

fig = px.histogram(concepcion_monthly_incidents, x='month_name', y='count', color='year',
              labels={'month_name':'Mes', 'count':'Número de focos de calor', 'year':'Año'},
              title='Número de incidentes de incendios mensual el Beni (2018-2025)')
fig.show()
```

```python
# calculamos en una tabla la ocurrencia de FC mensual
concepcion_monthly_incidents_sorted = concepcion_monthly_incidents.sort_values(by='count', ascending=False)

# Mostramos la tabla
display(concepcion_monthly_incidents_sorted)
```
