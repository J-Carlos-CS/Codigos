## 2. Primer Ejercicio

Codifica un programa en Python que nos permita guardar los nombres de los alumnos de una clase y las notas que han obtenido. Cada alumno puede tener distinta cantidad de notas. Guarda la informacin en un diccionario cuya claves sern los nombres de los alumnos y los valores sern listas con las notas de cada alumno. Al final el programa nos mostrar la lista de alumnos y la nota media obtenida por cada uno de ellos. Nota: si se introduce el nombre de un alumno que ya existe el programa nos deber generar un error.

```python
alumnos = {}

while(True):
  print("\n")
  print("1. Aadir/modificar un nuevo estudiante")
  print("2. Salir")
  opcion = int(input("Ingresa una opcin:"))

  if opcion == 1:
    alumno = input("Nombre del alumno:")
    while alumno in alumnos:
      print("Alumno ya existe.")
      alumno = input("Nombre del alumno:")

    notas=[]
    nota = int(input("Dame una nota del alumno (negativo para terminar):"))
    while nota >= 0:
      notas.append(nota)
      nota = int(input("Dame una nota del alumno (negative para terminar):"))
    alumnos[alumno] = notas.copy()
  else:
    break

for alumno, notas in alumnos.items():
  print("%s ha sacado de nota media %f"%(alumno,sum(notas)/len(notas)))
```

## 3. Segundo Ejercicio

En un notebook en COLAB:

Para una tienda de barrio, crear un programa para guardar los nombres, precios y cantidades vendidas de n productos durante los 12 meses del ao. Los valores de las ventas pueden ser generados de manera aleatoria. Al final, el programa deber brindar un menu para seleccionar entre las siguientes opciones:

- Debera mostrar el nombre y precio del producto que fue mas vendido.
- Acceder el nombre del producto, y obtener el total de ganancia generado por el mismo durante el ao. Si el producto no existe, deber mostrarnos un error.
- El producto menos requerido por los clientes.

Tras cada consulta el programa nos preguntar si queremos hacer otra consulta.

```python
import random

def generar_datos(num=10):
  productos = {}
  for i in range(num):
    nombre_producto = f"Producto {i+1}"
    precio = round(random.uniform(50,500),2)

    cantidades = []
    for _ in range(12):
      cantidades.append(random.randint(0,500))

    productos[nombre_producto] = {"precio":precio, "cantidades":cantidades}

  return productos

n = 5 # Definir cantidad de productos
productos = generar_datos(n)

def mas_vendido(productos):

  unidades_vendidas = []
  producto_maximo = ""
  cantidad_maxima = 0
  for nombre, datos in productos.items():
    unidades_vendidas = sum(datos["cantidades"])
    if unidades_vendidas > cantidad_maxima:
      cantidad_maxima = unidades_vendidas
      producto_maximo = nombre

  print(producto_maximo, productos[producto_maximo]["precio"])

while True:
  print("Menu")
  print("1. Nombre y precio del producto que fue mas vendido.")
  print("2. Acceder el nombre del producto, y obtener el total de ganancia generado por el mismo durante el ao.")
  print("3. El producto menos requerido por los clientes.")
  print("4. Salir.")

  opcion = int(input("Ingresa una opcin:"))

  if opcion == 1:
    mas_vendido(productos)
  elif opcion == 2:
    pass
  elif opcion == 3:
    pass
  else:
    break
```
