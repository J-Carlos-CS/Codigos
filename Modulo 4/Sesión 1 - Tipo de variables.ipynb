### Variables tipo int

`type(var1)`

### Variables tipo float



```python
a1 = 10.5
b1 = 4.0
c1 = 3.5

print("el tipo de variable para a1, es ", type(a1) )
print("el valor de la variable b1 es ", b1)
```

### Variables tipo boolean

Este tipo de variables toman solamente 2 valores:
`True`: Verdadera
`False`: Falso

```python
a2 = True
b2 = False
print("el tipo de variable para a2 es:",type(a2))
print("el valor de la variable b2 es:", b2)
```

### Variables tipo string
Son cadenas de caracteres que estan encerrados entre comillas simples (o debles)

```python
a2 = 'cadena'
a3 = 'a3'
print("el tipo de variable de a2",type(a2))
print("el valor de la variable a3:",a3)
```

```python
# esta linea no es ejecutada
# diferencia entre bool y strin
a4 = True
a5 = "True" # no es utilizado
print("el tipo de variable de a4 es:", type(a4))
print("el tipo de variable de a5 es:", type(a5))
```

Para cualquier variable, es posible obtener su refencia en memoria:


```python
# para la variable a5
print("el valor de la variable a5", a5)
print("su referencia en memoria (id) para a5", id(a5))
print("el tipo de variable para a5:",type(a5))
```

## Operadores aritmeticos

```python
# operado modulo o residuo `%`
# 7/3 = 3*2 + 1
# 8/3 = 3*2 + 2
print("residuo de dividir 7 entre 3", 7%3)
print("residuo de dividir 8 entre 3", 8%3)
```

```python
# operado de exponente
# 7*7 = 7^2 => 7**2
print("7 al cuadrado es", 7**2)
```

```python
# operador de division entera
# 5/3  = 1.666
# 5//3 = 1
print("division entera de 5 entre 3:", 5//3)

# -5/3 = -1.666
# -5//3 = -2
# -3 -2          -1
print("division entera de -5 entre 3:", -5//3)
```

### Operadores de comparacion

```python
# mayor igual que
a1 = 3
a2 = 5
b1 = a1 >= a2
print("el valor de b1 es",b1)
```

```python
# menor igual que
b2 = a1<=a2
print("el valor de b2 es:",b2)
```

```python
# igual a
b3 = a1 == a2
print("el valor de b3 es:",b3)
```

```python
# no es igual a
b4 = a1 != a2
print("el valor de b4 es:", b4)
```
