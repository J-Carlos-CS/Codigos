```python
# Definicion de listas
d = [0.5, 0.8, 0.9, 0.95 , 0.1, 0.3, 0.4]
# Encontrar una lista con los elementos que tengan un valor > 0.79
d_1 =[]
for elemento in d:
    if elemento >0.79:
        d_1.append(elemento)

print(d_1)
```

```python
# para cambiar de fraccion a porcentaje:
d_2 =[]
for elemento in d:
    elem2 = elemento*100
    d_2.append(elem2)
print(d_2)
```

```python
# ejemplo:
# para un modelo de regresion lineal:
c = [5, 0.2, 0.03, 0.01]
# y = c[0]+ c[1]*x + c[2]*x^2 + c[3]*x^3
x = [1, 1.5 , 3, 4]
# encontrar x^2
x_sq = []
for elemento in x:
    x2 = elemento*elemento
    x_sq.append(x2)
print(x)
print(x_sq)
```

```python
# Utilizando sintaxis pythonico
# compresion de listas
x = [1, 1.5 , 3, 4]
x_sq2 = [var*var for var in x]
print(x)
print(x_sq2)
```

```python
# comprension de listas con condicional if:
d = [0.5,0.8, 0.9 ,0.95 ,0.1 , 0.4]
# obtener una lista con los elementos cuyo valor > 0.79
d_1 =[]
for elemento in d:
    if elemento >0.79:
        d_1.append(elemento)
print(d_1)
```

```python
d_1p =[elem for elem in d if (elem>0.79)]
print(d_1p)
```

```python
# Comprension de listas con if y else
a = [5, -9, 7, -4, -2, 10]
# b = [P , N, ...]
b = ["Positivo" if num>0 else "Negativo" for num in a ]
print(a)
print(b)
```

```python
# sin compresion de listas
b_0 = []
for num in a:
    if num>0:
        b_0.append("Positivo")
    else:
        b_0.append("Negativo")
print(b_0)
```

```python
c = [num > 1 for num in a]
print(c)
```

## Compresion de diccionarios

```python
d1 = {"modelo1":0.95, "modelo2":0.7, "modelo3":0.95, "modelo4":0.3, "modelo5":0.1}
# obtener un diccionario que nos ayude a encontrar los modelos con precision mayor a 0.69
d2 = {clave:valor for clave,valor in d1.items() if valor>0.69}
print(d2)
```

```python
d2_0 = {}
for clave,valor in d1.items():
    if valor>0.69:
        d2_0[clave]=valor
print(d2_0)
```
