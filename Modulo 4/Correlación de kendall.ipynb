```python
from scipy import stats

# {A : 1,  B : 2,  C : 3,  D : 4}

ranking1 = [1, 2, 3, 4]
ranking2 = [1, 2, 3, 4]
resultado = stats.kendalltau(ranking1, ranking2)
print("Totalmente de acuerdo:", resultado.correlation)

ranking1 = [1, 2, 3, 4]
ranking2 = [4, 3, 2, 1]
resultado = stats.kendalltau(ranking1, ranking2)
print("Totalmente en desacuerdo:", resultado.correlation)

ranking1 = [1, 3, 2, 4]
ranking2 = [1, 4, 3, 2]
resultado = stats.kendalltau(ranking1, ranking2)
print("Parcialmente de acuerdo:", resultado.correlation)
```

| Caso             | Ranking 1 | Ranking 2 | τ Kendall | Interpretación                     |
| ---------------- | --------- | --------- | --------- | ---------------------------------- |
| Acuerdo total    | A B C D   | A B C D   | +1.0      | **Mismo orden**                    |
| Desacuerdo total | A B C D   | D C B A   | –1.0      | **Orden totalmente invertido**     |
| Acuerdo parcial  | A C B D   | A D C B   | ~0.33     | **Orden parcialmente coincidente** |
