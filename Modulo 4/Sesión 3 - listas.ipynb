El metodo `copy()`

```python
# Aliasing colocamos un "alias"
a = [10,20 ,30, 40, 50]
# Utilizamos el operador asignacion
b = a
print("lista a:", a)
print("lista b:", b)
```

```python
a[0] = 70
print("lista a:", a)
print("lista b:", b)
```

```python
b[0] = -55
print("lista a:", a)
print("lista b:", b)
```

```python
a = [10,20,30,40,50]
# para evitar el problema de aliasing
b = a.copy()
a[0] = 70
b[0] =-55
print("lista a:", a)
print("lista b:", b)
```

## Bucles / Loops

```python
a = [3,6,7,9]
for var in a:
  print("el valor de var es:",var)
```

```python
for var1 in range(10):
  # en el primer bucle var1 = 0
  print("el valor de var1:",var1)
```

```python
for var1 in range(10):
  # en el primer bucle var1 = 0
  print("el valor de var1:",var1)
  # var2 = 0 * 10 = 0
  var2 = var1*10
  print("el valor de var2:",var2)
```

```python
# bucles anidados
a = [1,2,3]
b = [4,5,6]
for var1 in a:
  for var2 in b:
    print(var1, var2, var1+var2)
  print("----")
```

```python
a = 7

while (a > 2 ):
  print("la condicion es verdadera")
  a = a - 1
  print("el valor de la condicion:", a>2)
  print("-----")
```

```python
b1 = "abc"
print(len(b1))
```

```python
while len(b1)<10:
  b1 = b1+"a"
  print(b1)
```

```python
lista_1 = [5,7,9]
sum(lista_1)
```

## Diccionarios en python:


```python
dict_1 = { "Peru":"Lima", "Ecuador":"Quito", "Colombia":"Bogota"  }
```

```python
print(dict_1)
```

```python
# Las claves(keys) son inmutables (no pueden cambiar su valor una vez que son definidos)
# Los valores(values) son mutables (sus valores pueden ser cambiados)
print(dict_1["Peru"])
dict_1["Peru"] = ["Lima", "Cuzco"]
print(dict_1)
```

```python
dict_1["Ecuador"]= ["Quito","Guayaquil"]
print(dict_1)
```

```python
dict_1["Colombia"] = ["Bogota","Barranquilla"]
print(dict_1)
```
