**Inspeccion Inicial**

```python
# ==============================
# üìä ESTAD√çSTICAS DESCRIPTIVAS B√ÅSICAS
# ==============================

# 1Ô∏è‚É£ Contar n√∫mero de registros
num_registros = len(df)
print(f"üìã N√∫mero de registros: {num_registros}")

# 2Ô∏è‚É£ Calcular media y desviaci√≥n est√°ndar
media_edad = df["Edad"].mean()
desv_edad = df["Edad"].std()

media_ingreso = df["Ingreso_Mensual"].mean()
desv_ingreso = df["Ingreso_Mensual"].std()

media_prestamo = df["Monto_Prestamo"].mean()
desv_prestamo = df["Monto_Prestamo"].std()

# 3Ô∏è‚É£ Mostrar resultados
print("\nüìà Estad√≠sticas descriptivas:")
print(f"Edad -> Media: {media_edad:.2f} | Desviaci√≥n est√°ndar: {desv_edad:.2f}")
print(f"Ingreso_Mensual -> Media: {media_ingreso:.2f} | Desviaci√≥n est√°ndar: {desv_ingreso:.2f}")
print(f"Monto_Prestamo -> Media: {media_prestamo:.2f} | Desviaci√≥n est√°ndar: {desv_prestamo:.2f}")

# 4Ô∏è‚É£ (Opcional) Ver todo el resumen estad√≠stico con describe()
print("\nüîç Resumen completo:")
display(df[["Edad","Ingreso_Mensual","Monto_Prestamo"]].describe())
```

**Identifiaci√≥n de errores**

```python
# ==============================
# üîé VERIFICACIONES DE CALIDAD DE DATOS (Colab)
# Requisitos: df con columnas:
# ID_Cliente, Edad, Ingreso_Mensual, Monto_Prestamo, Sexo, Region
# ==============================
import pandas as pd
import numpy as np
from IPython.display import display

# 1) Edades negativas o mayores a 100
edad_invalida = (df["Edad"] < 0) | (df["Edad"] > 100)
print(f"üë∂ Edades inv√°lidas (negativas o >100): {edad_invalida.sum()}")
display(df.loc[edad_invalida])

# 2) Duplicados
# 2a) Duplicados por ID_Cliente
dup_por_id = df.duplicated(subset=["ID_Cliente"], keep=False)
print(f"ü™™ Duplicados por ID_Cliente: {dup_por_id.sum()}")
display(df.loc[dup_por_id].sort_values("ID_Cliente"))

# 2b) Duplicados por combinaci√≥n de campos (sin usar ID)
campos_combo = ["Edad","Ingreso_Mensual","Monto_Prestamo","Sexo","Regi√≥n"]
dup_por_combo = df.duplicated(subset=campos_combo, keep=False)
print(f"üß© Duplicados por combinaci√≥n {campos_combo}: {dup_por_combo.sum()}")
display(df.loc[dup_por_combo].sort_values(campos_combo))

# 3) Casos con Monto_Prestamo > Ingreso_Mensual
prestamo_mayor_que_ingreso = df["Monto_Prestamo"] > df["Ingreso_Mensual"]
print(f"üí∏ Casos con pr√©stamo > ingreso: {prestamo_mayor_que_ingreso.sum()}")
display(df.loc[prestamo_mayor_que_ingreso, ["ID_Cliente","Ingreso_Mensual","Monto_Prestamo","Sexo","Regi√≥n","Edad"]])

# 4) Ingresos o montos iguales a 0
ingresos_o_montos_cero = (df["Ingreso_Mensual"] == 0) | (df["Monto_Prestamo"] == 0)
print(f"üßÆ Registros con ingreso=0 o monto=0: {ingresos_o_montos_cero.sum()}")
display(df.loc[ingresos_o_montos_cero, ["ID_Cliente","Ingreso_Mensual","Monto_Prestamo","Sexo","Regi√≥n","Edad"]])

# 5) Inconsistencias de texto en Regi√≥n (p.ej. 'La Paz', 'la paz', 'LA PAZ')
#    Normalizaci√≥n: quitar espacios, pasar a min√∫sculas y remover acentos
def normalizar_texto(s):
    if pd.isna(s):
        return s
    s = str(s).strip().lower()
    # remover acentos
    import unicodedata
    s = ''.join(c for c in unicodedata.normalize('NFKD', s) if not unicodedata.combining(c))
    return s

region_norm = df["Regi√≥n"].apply(normalizar_texto)
df["_Region_norm"] = region_norm

# Mostrar variantes originales por forma normalizada
variantes = (
    df.groupby("_Region_norm")["Regi√≥n"]
      .apply(lambda x: sorted(x.unique().tolist()))
      .rename("Variantes_Originales")
      .reset_index()
)
print("üåç Variantes de 'Regi√≥n' por forma normalizada:")
display(variantes)

# (Opcional) Tambi√©n revisar inconsistencias en 'Sexo'
sexo_norm = df["Sexo"].apply(normalizar_texto)
df["_Sexo_norm"] = sexo_norm
variantes_sexo = (
    df.groupby("_Sexo_norm")["Sexo"]
      .apply(lambda x: sorted(x.unique().tolist()))
      .rename("Variantes_Originales")
      .reset_index()
)
print("üë§ Variantes de 'Sexo' por forma normalizada:")
display(variantes_sexo)

# ==============================
# üìã Resumen de banderas por registro (para auditor√≠a)
# ==============================
df_flags = pd.DataFrame({
    "ID_Cliente": df["ID_Cliente"],
    "edad_invalida": edad_invalida,
    "dup_por_id": dup_por_id,
    "dup_por_combo": dup_por_combo,
    "prestamo_mayor_que_ingreso": prestamo_mayor_que_ingreso,
    "ingreso_o_monto_cero": ingresos_o_montos_cero
})

print("üìä Conteo total de banderas:")
display(df_flags.drop(columns=["ID_Cliente"]).sum().to_frame("conteo"))

print("üßæ Primeros registros con sus banderas (muestra):")
display(df_flags.head(10))
```

**Limpieza B√°sica**

```python
# ==============================
# üßπ LIMPIEZA B√ÅSICA DE DATOS
# ==============================

# 1Ô∏è‚É£ Crear una copia del DataFrame original
df_limpio = df.copy()

# 2Ô∏è‚É£ Corregir o eliminar registros con errores
# - Edades negativas o mayores a 100 ‚Üí se reemplazan por NaN (faltantes)
df_limpio.loc[(df_limpio["Edad"] < 0) | (df_limpio["Edad"] > 100), "Edad"] = None

# - Ingreso o monto igual a 0 ‚Üí marcar como NaN
df_limpio.loc[df_limpio["Ingreso_Mensual"] <= 0, "Ingreso_Mensual"] = None
df_limpio.loc[df_limpio["Monto_Prestamo"] <= 0, "Monto_Prestamo"] = None

# - Casos donde el monto del pr√©stamo es mayor al ingreso ‚Üí marcar el monto como NaN
df_limpio.loc[df_limpio["Monto_Prestamo"] > df_limpio["Ingreso_Mensual"], "Monto_Prestamo"] = None

# 3Ô∏è‚É£ Normalizar texto
# Pasar ‚ÄúSexo‚Äù y ‚ÄúRegi√≥n‚Äù a may√∫scula inicial (ej. "Femenino", "La Paz")
df_limpio["Sexo"] = df_limpio["Sexo"].str.strip().str.capitalize()
df_limpio["Regi√≥n"] = df_limpio["Regi√≥n"].str.strip().str.title()

# 4Ô∏è‚É£ Quitar duplicados (mismos valores en todas las columnas relevantes)
df_limpio = df_limpio.drop_duplicates(subset=["Edad","Ingreso_Mensual","Monto_Prestamo","Sexo","Regi√≥n"])

# ==============================
# üìä RE-C√ÅLCULO DE ESTAD√çSTICAS
# ==============================
media_edad_limpia = df_limpio["Edad"].mean()
desv_edad_limpia = df_limpio["Edad"].std()

media_ingreso_limpia = df_limpio["Ingreso_Mensual"].mean()
desv_ingreso_limpia = df_limpio["Ingreso_Mensual"].std()

media_prestamo_limpia = df_limpio["Monto_Prestamo"].mean()
desv_prestamo_limpia = df_limpio["Monto_Prestamo"].std()

print("‚úÖ Dataset limpio generado.\n")
print(f"Edad -> Media: {media_edad_limpia:.2f} | Desviaci√≥n est√°ndar: {desv_edad_limpia:.2f}")
print(f"Ingreso_Mensual -> Media: {media_ingreso_limpia:.2f} | Desviaci√≥n est√°ndar: {desv_ingreso_limpia:.2f}")
print(f"Monto_Prestamo -> Media: {media_prestamo_limpia:.2f} | Desviaci√≥n est√°ndar: {desv_prestamo_limpia:.2f}")

# ==============================
# üîç Comparaci√≥n antes vs. despu√©s
# ==============================
print("\nüìà Comparaci√≥n antes vs. despu√©s de la limpieza:")
comparacion = pd.DataFrame({
    "Variable": ["Edad","Ingreso_Mensual","Monto_Prestamo"],
    "Media_Antes": [
        df["Edad"].mean(),
        df["Ingreso_Mensual"].mean(),
        df["Monto_Prestamo"].mean()
    ],
    "Media_Despu√©s": [
        media_edad_limpia,
        media_ingreso_limpia,
        media_prestamo_limpia
    ],
    "Desv.Antes": [
        df["Edad"].std(),
        df["Ingreso_Mensual"].std(),
        df["Monto_Prestamo"].std()
    ],
    "Desv.Despu√©s": [
        desv_edad_limpia,
        desv_ingreso_limpia,
        desv_prestamo_limpia
    ]
})
display(comparacion)

# 5Ô∏è‚É£ Mostrar muestra de los datos limpios
print("\nüßæ Vista previa del dataset limpio:")
display(df_limpio.head())
```
