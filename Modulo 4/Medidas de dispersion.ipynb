```python
import numpy as np

# Contexto: levantamiento de pesas
pesos = [5, 15, 25, 35, 40, 55, 80, 90, 170]

print("Rango:", np.max(pesos) - np.min(pesos))

print("Varianza poblacional:", np.var(pesos))
print("Varianza muestral:", np.var(pesos, ddof=1))

print("Desviaci√≥n est√°ndar poblacional:", np.std(pesos))
print("Desviaci√≥n est√°ndar muestral:", np.std(pesos, ddof=1))

#C√°lculo del rango intercuart√≠lico
Q1 = np.percentile(pesos, 25)
Q3 = np.percentile(pesos, 75)
print("Rango intercuart√≠lico:", Q3 - Q1)
```

```python
Interpretaci√≥n:
La diferencia entre el levantamiento m√°s bajo y el m√°s alto es muy grande (165 kg).
Esto indica que en el grupo hay personas con niveles muy diferentes de fuerza ‚Äî desde principiantes hasta atletas avanzados.
```

1) Rango: 165

El rango representa la diferencia entre el peso m√°ximo y m√≠nimo levantado.

‚û° Interpretaci√≥n:
Existe una diferencia muy grande (165 kg) entre el levantamiento m√°s bajo y el m√°s alto.
Esto indica que en el grupo hay gran diversidad de niveles de fuerza, desde principiantes (5‚Äì15 kg) hasta atletas muy avanzados (90‚Äì170 kg).

```python
# --- Estad√≠sticos descriptivos y gr√°ficos de levantamiento de pesas ---
import numpy as np
import matplotlib.pyplot as plt

# Datos: levantamiento de pesas (kg)
pesos = [5, 15, 25, 35, 40, 55, 80, 90, 170]

# C√°lculos b√°sicos
rango = np.max(pesos) - np.min(pesos)
var_poblacional = np.var(pesos)
var_muestral = np.var(pesos, ddof=1)
std_poblacional = np.std(pesos)
std_muestral = np.std(pesos, ddof=1)
Q1 = np.percentile(pesos, 25)
Q3 = np.percentile(pesos, 75)
RIC = Q3 - Q1

# --- Imprimir resultados en consola ---
print("üìä Resultados estad√≠sticos")
print(f"Rango: {rango}")
print(f"Varianza poblacional: {var_poblacional:.2f}")
print(f"Varianza muestral: {var_muestral:.2f}")
print(f"Desviaci√≥n est√°ndar poblacional: {std_poblacional:.2f}")
print(f"Desviaci√≥n est√°ndar muestral: {std_muestral:.2f}")
print(f"Rango intercuart√≠lico (IQR): {RIC:.2f}")

# --- Gr√°ficas ---
fig, axes = plt.subplots(1, 2, figsize=(12,5))

# Histograma
axes[0].hist(pesos, bins=6, color="skyblue", edgecolor="black")
axes[0].set_title("Distribuci√≥n de Pesos (kg)")
axes[0].set_xlabel("Peso (kg)")
axes[0].set_ylabel("Frecuencia")

# Boxplot
axes[1].boxplot(pesos, vert=False, patch_artist=True,
                boxprops=dict(facecolor="lightgreen", color="black"),
                medianprops=dict(color="red"))
axes[1].set_title("Boxplot de Pesos")
axes[1].set_xlabel("Peso (kg)")

# A√±adir valores como nota bajo las gr√°ficas
metrics_text = (f"Rango: {rango} | Varianza poblacional: {var_poblacional:.2f} | "
                f"Varianza muestral: {var_muestral:.2f} | "
                f"Desv. poblacional: {std_poblacional:.2f} | "
                f"Desv. muestral: {std_muestral:.2f} | "
                f"IQR: {RIC:.2f}")

plt.figtext(0.5, -0.05, metrics_text, ha="center", fontsize=10,
            bbox=dict(facecolor="white", alpha=0.7))

plt.tight_layout()
plt.show()
```
