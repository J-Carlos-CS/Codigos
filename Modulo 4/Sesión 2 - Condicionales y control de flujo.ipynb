```python
# La condicion: Un valor booleano (True o False)
a1 = 5
a2 = 7
condicion = a1>a2
print(condicion)
```

```python
a3 = 10
if (a1>a2):
  a3 = a3+5
  print("la condicion es verdadera")
else:
  a3 = a3 - 5
  print("la condicion es falsa")

print("el valor de a3 es:",a3)

```

```python
condicion1 = True
condicion2 = False

if (condicion1 or condicion2):
  print("la condicion total es verdadera")

if (condicion1 and condicion2):
  print("la condicion total2 es verdadera")
else:
  print("la condicion total2 es falsa")
```

```python
# acceso a elementos de una lista
#              0            1        2       3           4      5          6
frutas = ['mandarina', 'manzana', 'pera', 'plátano', 'kiwi', 'manzana', 'plátano']
#             -7            -6       -5      -4        -3      -2          -1

# para acceder al ultimo elemento
print("el ultimo elemento",frutas[6])
print("el ultimo elemento",frutas[-1])
print("el TERCER elemento",frutas[-5])
print("el elementos",frutas[2:6])
```

```python
# para acceder a multiples elementos
print("los elementos 1 al 5", frutas[0:5])
```

```python
# para acceder a multiples elementos con salto
print("los elementos 1 al 5", frutas[0:5:2])
```

```python
# el metodo len: para obtener el numero de elementos de la lista
print("el numero de elementos es:",len(frutas))
len(frutas)
```
